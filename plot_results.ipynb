```python
import matplotlib.pyplot as plt

epochs = [1, 2, 3, 4]
batch_steps = [1550, 3100, 4650, 6200, 7750]

batch_losses = {
    1: [0.4116619885544623, 0.26802527021496525, 0.2107396884118357, 0.1823147337883711, 0.1638283923436557],
    2: [0.14870520848901042, 0.1435529226268972, 0.13018675905081534, 0.12635745445326452, 0.12180359318251571],
    3: [0.11513394294727233, 0.1105682559876192, 0.10822313634318209, 0.10339748898881577, 0.10223214082209574],
    4: [0.09623086175971454, 0.09266305069499199, 0.09140441183962168, 0.09412516790712552, 0.09113986231687088]
}

train_losses = [0.1638283923436557, 0.12180359318251571, 0.10223214082209574, 0.09113986231687088]
valid_losses = [0.14247334521799865, 0.12233895088088242, 0.1088451296374463, 0.08133085497836051]
accuracies = [0.9464297890663147, 0.9529033899307251, 0.9590383768081665, 0.97023606300354]

plt.figure(figsize=(12, 6))
for epoch in epochs:
    plt.plot(batch_steps, batch_losses[epoch], marker='o', label=f'Epoch {epoch}')
plt.title('Batch Losses per Epoch')
plt.xlabel('Batch Steps')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()

fig, ax1 = plt.subplots(figsize=(12, 6))

ax1.plot(epochs, train_losses, marker='o', label='Train Loss', color='blue')
ax1.plot(epochs, valid_losses, marker='o', label='Validation Loss', color='orange')
ax1.set_xlabel('Epochs')
ax1.set_ylabel('Loss')
ax1.set_title('Training and Validation Losses and Accuracy')
ax1.legend(loc='upper left')
ax1.grid(True)

ax2 = ax1.twinx()
ax2.plot(epochs, accuracies, marker='o', label='Accuracy', color='green')
ax2.set_ylabel('Accuracy')
ax2.legend(loc='upper right')

plt.show()
```
